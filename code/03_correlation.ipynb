```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read data
YouGov = pd.read_csv("../data/cleaned_YouGov.csv")
YouGov["endtime"] = pd.to_datetime(YouGov["endtime"])
```

```python
# Compute correlations with face mask behaviour
target1 = 'face_mask_behaviour_binary'

df_mask = YouGov.drop(columns=['protective_behaviour_binary'])
corr_mask = df_mask.corr()['face_mask_behaviour_binary'].sort_values(ascending=False)

print("correlations (Mask):")
print(corr_mask)
```

```python
plt.figure(figsize=(6, 12))
sns.heatmap(corr_mask.to_frame(), cmap='coolwarm')
plt.title('Correlation with mask Wearing')
plt.show()
```

```python
# Compute correlations with overall protective behaviour
target2 = 'protective_behaviour_binary'

df_protect = YouGov.drop(columns=['face_mask_behaviour_binary', 'protective_behaviour_nomask_scale'])
corr_protect = df_protect.corr()['protective_behaviour_binary'].sort_values(ascending=False)

print("Top positive correlations (Protective):")
print(corr_protect)
```

```python
plt.figure(figsize=(6, 12))
sns.heatmap(corr_protect.to_frame(), cmap='coolwarm')
plt.title('Correlation with protective behaviour')
plt.show()
```
