```python
import pandas as pd
from sklearn.model_selection import train_test_split

before_df = pd.read_csv("../data/before_mandate_data.csv")
after_df = pd.read_csv("../data/after_mandate_data.csv")
```

```python
# before period

before_df = before_df.drop(columns=["protective_behaviour_binary"])

y_before = before_df["face_mask_behaviour_binary"]
X_before = before_df.drop(columns=["face_mask_behaviour_binary"])

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_before, y_before,
    test_size=0.2,
    random_state=3034,
    stratify=y_before
)

X_train.to_csv("../train_test/before_train_X_face_mask.csv", index=False)
X_test.to_csv("../train_test/before_test_X_face_mask.csv", index=False)
y_train.to_csv("../train_test/before_train_y_face_mask.csv", index=False)
y_test.to_csv("../train_test/before_test_y_face_mask.csv", index=False)
```

```python
# after period
after_df = after_df.drop(columns=["protective_behaviour_binary"])

y_after = after_df["face_mask_behaviour_binary"]
X_after = after_df.drop(columns=["face_mask_behaviour_binary"])

X_after_train, X_after_test, y_after_train, y_after_test = train_test_split(
    X_after, y_after,
    test_size=0.2,
    random_state=3034,
    stratify=y_after
)

X_after_train.to_csv("../train_test/after_train_X_face_mask.csv", index=False)
X_after_test.to_csv("../train_test/after_test_X_face_mask.csv", index=False)
y_after_train.to_csv("../train_test/after_train_y_face_mask.csv", index=False)
y_after_test.to_csv("../train_test/after_test_y_face_mask.csv", index=False)
```
