```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Random Forest feature importance comparison for face mask behaviour prediction

# Read data
before_df = pd.read_csv("../importance/rf_before_face_mask_importance.csv")
after_df = pd.read_csv("../importance/rf_after_face_mask_importance.csv")

# Select top 10 features from each period
before_top10 = before_df.sort_values("median_importance", ascending=False).head(10)
after_top10 = after_df.sort_values("median_importance", ascending=False).head(10)

# Merge feature sets
features = list(set(before_top10["feature"]) | set(after_top10["feature"]))

# Align data and fill missing values with 0
before_plot = before_top10.set_index("feature").reindex(features)
after_plot = after_top10.set_index("feature").reindex(features)

before_values = before_plot["median_importance"].fillna(0)
after_values = after_plot["median_importance"].fillna(0)

before_q1 = before_plot["q1"].fillna(0)
before_q3 = before_plot["q3"].fillna(0)

after_q1 = after_plot["q1"].fillna(0)
after_q3 = after_plot["q3"].fillna(0)

# Sort features by overall importance
order = (
    pd.DataFrame({
        "feature": features,
        "max_importance": np.maximum(before_values, after_values)
    })
    .sort_values("max_importance", ascending=False)["feature"]
)

before_values = before_values.reindex(order)
after_values = after_values.reindex(order)

before_q1 = before_q1.reindex(order)
before_q3 = before_q3.reindex(order)

after_q1 = after_q1.reindex(order)
after_q3 = after_q3.reindex(order)

# Error bars
before_xerr = [before_values - before_q1, before_q3 - before_values]
after_xerr = [after_values - after_q1, after_q3 - after_values]

# Plot mirrored bar chart
y_pos = np.arange(len(order))

plt.figure(figsize=(10, 6))

plt.barh(y_pos, -before_values, xerr=before_xerr, color="#1f77b4", alpha=0.85, capsize=3)
plt.barh(y_pos,  after_values, xerr=after_xerr,  color="#ff0e0e", alpha=0.85, capsize=3)


plt.yticks(y_pos, order, fontsize=12)
plt.gca().invert_yaxis()
plt.axvline(0, color="black", linewidth=1)

# Format x-axis to show positive values on both sides
max_val = max(before_values.max(), after_values.max())
plt.xlim(-max_val * 1, max_val * 1)
ticks = plt.xticks()[0]
plt.xticks(ticks, [f"{abs(t):.3f}" for t in ticks], fontsize=12)

plt.xlabel("Median feature importance", fontsize=12)
plt.grid(axis="x", alpha=0.25)

plt.tight_layout()
plt.show()
```
